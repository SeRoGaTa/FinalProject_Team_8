```python
import pandas as pd
import numpy as np
from selenium import webdriver
import time
import ssl
import wikipedia
from webdriver_manager.chrome import ChromeDriverManager

ssl._create_default_https_context = ssl._create_unverified_context
```

```python
weather = pd.read_csv('races.csv')
weather = weather.loc[(weather['year'] >= 2018) & (weather['year'] < 2022)]
weather['url']= weather['url'].replace('São_Paulo_Grand_Prix','2021_São_Paulo_Grand_Prix', regex=True)
weather
```

```python
info = []

for link in weather.url:
    try:
        df = pd.read_html(link)[0]
        if 'Wear' in list(df.iloc[:,0]):
            n = list(df.iloc[:,0]).index('Weather')
            info.append(df.iloc[n,1])
        else:
            driver = webdriver.Chrome(ChromeDriverManager().install())
            driver.get(link)

            # click language button
            time.sleep(2)
            button = driver.find_element_by_xpath("//button[@class='mw-interlanguage-selector mw-ui-button']")
            button.click()
            time.sleep(2)
            button = driver.find_element_by_xpath("//a[@lang='it']").get_attribute('href')

            driver2 = webdriver.Chrome(ChromeDriverManager().install())
            driver2.get(button)
            driver.quit()
                        
            # find weather in italian with selenium
                        
            clima = driver2.find_element_by_xpath('//*[@id="mw-content-text"]/div/table[1]/tbody/tr[9]/td').text
            info.append(clima)
            driver2.quit()
    except:
        info.append('not found')
```

```python
weather['weather'] = info
```

```python
# Export to a csv.
weather.to_csv("Formula1_weatherbase.csv", index=False)
```
